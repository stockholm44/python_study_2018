# 데이터 탐색 및 시각화

- ID : 집을 구분하는 번호
- date : 집을 구매한 날짜
- price : 집의 가격(Target variable)
- bedrooms : 침실의 수
- bathrooms : 침실 개수 당 화장실의 수(화장실의 수 / 침실의 수 )
- sqft_living : 주거 공간의 평방 피트(면적)
- sqft_lot : 부지의 평방 피트(면적)
- floors : 집의 층 수
- waterfront : 집의 전방에 강이 흐르는지 유무 (a.k.a. 리버뷰)
- view : 집이 얼마나 좋아 보이는지의 정도
- condition : 집의 전반적인 상태
- grade : King County grading 시스템 기준으로 매긴 집의 등급
- sqft_above : 지하실을 제외한 평방 피트(면적)
- sqft_basement : 지하실의 평방 피트(면적)
- yr_built : 지어진 년도
- yr_renovated : 집을 재건축한 년도
- zipcode : 우편번호
- lat : 위도
- long : 경도
- sqft_living15 : 2015년 기준 주거 공간의 평방 피트(면적, 집을 재건축했다면, 변화가 있을 수 있음)
- sqft_lot15 : 2015년 기준 부지의 평방 피트(면적, 집을 재건축했다면, 변화가 있을 수 있음)

```python
# date의 진짜 date부분만 남기고 나머지는 날리자. 그담에 년,월,일,요일까지 뽑자.
train_df['date_parse'] =train_df['date'].map(lambda x : x[:8])
train_df.drop('date', axis=1, inplace=True)

# 년, 월, 일, 요일 순으로 Column 만들기
train_df['buy_year'] = train_df['date_parse'].map(lambda x : x[:4])
train_df['buy_month'] = train_df['date_parse'].map(lambda x : x[4:6])
train_df['buy_day'] = train_df['date_parse'].map(lambda x : x[6:8])
train_df['buy_day_of_week'] = pd.to_datetime(train_df['date_parse'], format='%Y%m%d').dt.weekday_name
```

```python
sns.distplot(train_df['price'] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(train_df['price'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train_df['price'], plot=plt)
plt.show()
```

```python
#We use the numpy fuction log1p which  applies log(1+x) to all elements of the column
train_df["price_log"] = np.log1p(train_df["price"])

#Check the new distribution 
sns.distplot(train_df['price_log'] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(train_df['price_log'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(train_df['price_log'], plot=plt)
plt.show()
```

## 추가로 만들 Feature
1. 전체 넓이(sqft_total) : sqft_above+sqft_basement 
2. 전체 넓이중 주거공간의 면적 비율 : sqft_living/sqft_total 
3. 한개층 면적(sqft_1st) : sqft_above/floor 
4. yr_renovated : 여기에 0 값들을 yr_built로 채움.
5. 위도경도를 이용해서 도심으로부터의 거리 계산

```python
train_df['sqft_total'] = train_df['sqft_above'] + train_df['sqft_basement']
train_df['living_ratio'] = train_df['sqft_living']/train_df['sqft_total']
train_df['sqft_1st_floor'] = train_df['sqft_above']/train_df['floors']
train_df.loc[train_df['yr_renovated']== 0, 'yr_renovated'] = train_df['yr_built']
```

# 모델선택과 훈련

### 1. LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_df2, y_train_df, test_size=0.33, random_state=42)

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

y_train_pred = lin_reg.predict(X_train)
y_test_pred = lin_reg.predict(X_test)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(y_train, y_train_pred, c='r', label='X_train')
ax.scatter(y_test, y_test_pred, c='b', label='X_test')
ax.set_title('LinearRegression')
plt.show()
```

### 2. LinearRegression(with PolynomialFeatures)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(train_df2_scaler)
X_poly.shape

X_train, X_test, y_train, y_test = train_test_split(X_poly, y_train_df, test_size=0.33, random_state=42)

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

y_train_pred = lin_reg.predict(X_train)
y_test_pred = lin_reg.predict(X_test)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(y_train, y_train_pred, c='r', label='X_train')
ax.scatter(y_test, y_test_pred, c='b', label='X_test')
ax.set_title('LinearRegression(with PolynomialFeatures)')
plt.show()
```
