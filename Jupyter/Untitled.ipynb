# 꺾은선 그래프

꺾은선 그래프 그리기

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot([1, 3])

plt.show()
```

X값과 Y값을 지정한 꺾은 선 그래프

```python
fig = plt.figure()
ax = fig.add_subplot(111)

x = [0, 2, 4]
y = [0, 4, 2]

ax.plot(x, y)

plt.show()
```

여러개의 선을 그리는 경우

```python
fig = plt.figure()
ax = fig.add_subplot(111)

x = [0, 2, 4]
y1 = [0, 4, 2.5]
y2 = [4 ,0, 1.5]

ax.plot(x, y1)
ax.plot(x, y2)

plt.show()
```
