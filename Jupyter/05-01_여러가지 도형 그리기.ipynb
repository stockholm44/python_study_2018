# 05-01 여러가지 도형 그리기

원호그리기
- matplotlib.patches.Arc 클래스로 그림,
- xy에 튜플형 중심좌표
- width, height에 폭과 높이
- angle에 도수로 회전 각도
- theta1, theta2 에 시작각도와 종료각도

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Arc 원호
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

c1 = mpatches.Arc(xy=(0.4,0.5), width=0.4, height= 0.4, angle=30,
                 theta1=0, theta2=230, linewidth=10, color='lightblue')
c2 = mpatches.Arc(xy=(0.6,0.5), width=0.4, height= 0.4, angle=30,
                 theta1=0, theta2=230, linewidth=10, color='pink')

ax.add_patch(c1)
ax.add_patch(c2)

plt.show()
```

화살표 그리기
- maplotlib.patches.Arrow 클래스
- x, y에 시작점 x, y를
- dx, dy에 시작점좌표로부터의 폭과 높이
- width에 화살표의 폭 지정

```python
# Arrow 화살표
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

a1 = mpatches.Arrow(x=0.1, y=0.4, dx=0.6, dy=0.5, width=0.5, color='lightblue')
a2 = mpatches.Arrow(x=0.3, y=0.1, dx=0.6, dy=0.5, width=0.5, color='pink')
ax.add_patch(a1)
ax.add_patch(a2)

plt.show()
```

화살표 서식을 보다 상세하게 하고 싶을 때
- matplotlib.patches.FancyArrow 클래스 이용
- 지정항목을 추가하여 
- head_width(화살표 끝의 폭)
- head_length(화살표 끝 길이)
- shape(형태지정 지정가능) full, left, right

```python
# FacyArrow 화살표
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)
a1 = mpatches.FancyArrow(x=0.2, y=0.3, dx=0.4, dy=0.3, width=0.1,
                        head_width=0.3, head_length=0.2,
                         shape='right', color='lightblue')
a2 = mpatches.FancyArrow(x=0.2, y=0.3, dx=0.4, dy=0.3, width=0.1,
                        head_width=0.3, head_length=0.2,
                         shape='left', color='pink')

ax.add_patch(a1)
ax.add_patch(a2)

plt.show()
```

원형그리기
- matplotlib.patches.Circle
- xy에 튜플형 중심좌표
- radius 에 반지름

```python
# CIrcle 원형
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

c1 = mpatches.Circle(xy=(0.5,0.5), radius=0.4, color='lightblue')
c2 = mpatches.Circle(xy=(0.5,0.5), radius=0.2, color='white')
ax.add_patch(c1)
ax.add_patch(c2)

plt.show()
```

정다각형 그리기
- matplotlib.patches.CirclePolygon
- 원형과 인수가 비슷.
- resolution 추가.(변의수)

```python
# Circle Polygon 정다각형
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)
cp1 = mpatches.CirclePolygon(xy=(0.5,0.5), radius=0.4, resolution=6, color='lightblue')
cp2 = mpatches.CirclePolygon(xy=(0.5,0.5), radius=0.2, resolution=8, color='white')

ax.add_patch(cp1)
ax.add_patch(cp2)

plt.show()
```

타원형 그리기
- matplotlib.patches.Ellipse
- xy 중심좌표
- width, height 폭 넓이
- angle 회전각도

```python
# Ellipese 타원형
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

e1 = mpatches.Ellipse(xy=(0.5,0.5), width=0.8, height=0.5, angle=30, color='lightblue')
e2 = mpatches.Ellipse(xy=(0.5,0.5), width=0.5, height=0.2, angle=30, color='white')

ax.add_patch(e1)
ax.add_patch(e2)

plt.show()
```

부채꼴 그리기
- matplotlib.patches.Wedge
- center 중심좌펴
- r 반지름
- theta1, theta2 베이자국 시작점, 종료점 각도

```python
# Wedge 부채꼴
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

w1 = mpatches.Wedge(center=(0.3,0.55), r=0.25, theta1=30, theta2=-30, color='pink')
w2 = mpatches.Wedge(center=(0.7,0.4), r=0.25, theta1=210, theta2=150, color='lightblue')

ax.add_patch(w1)
ax.add_patch(w2)

plt.show()
```

직사각형 그리기
- matplotlib.patches.Rectangle
- xy 왼쪽 아래각의 좌표
- width, height 에 좌표로부터의 폭과 높이
- angle 회전각도

```python
# Rectangle 직사각형
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

r1 = mpatches.Rectangle(xy=(0.1,0.4), width=0.5, height=0.5, angle=0, color='lightblue')
r2 = mpatches.Rectangle(xy=(0.4,0.1), width=0.5, height=0.5, angle=0, color='pink')

ax.add_patch(r1)
ax.add_patch(r2)

plt.show()
```

다각형 그리기
- matplotlib.patches.Polygon
- xy 꼭지점좌표
- 경로를 포함한 다각형그릴수 있으나 자세한 내용은 튜토리얼 참고

```python
# Polygon 다각형
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

p1 = mpatches.Polygon(xy=[(0,0.2), (0.3,0.8), (0.6,0.2)], facecolor='lightblue')
p2 = mpatches.Polygon(xy=[(0.4,0.8), (0.7,0.2), (1.0,0.8)], facecolor='pink')

ax.add_patch(p1)
ax.add_patch(p2)

plt.show()
```
