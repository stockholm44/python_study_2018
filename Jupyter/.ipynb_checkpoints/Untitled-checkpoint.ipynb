```python
import matplotlib.pyplot as plt
```

### 피겨(figure)와 서브플롯(subplot)
- 피겨에 서브플롯을 추가하는 방법으로 2가지가 있음
 - Figure.add_subplot()
 - pyplot.subplots()

add_subplot() 메서드로 서브플롯 배치하기

```python
# 피겨 생성
fig = plt.figure()

# 피겨 안에 서브플롯을 3개 배치한다.
ax1 = fig.add_subplot(221) # 2행2열 1번
ax2 = fig.add_subplot(222) # 2행2열 2번
ax3 = fig.add_subplot(223) # 2행2열 3번

plt.show()
```

```python
# 서브플롯 번호 할당 순서의 확인.
fig = plt.figure()

# 서브를롯 작성
ax1 = fig.add_subplot(221) 
ax2 = fig.add_subplot(222) 
ax3 = fig.add_subplot(223) 

# 번호 기술
for i, ax in enumerate([ax1, ax2, ax3], start=1):
    txt = 'ax{0}\n(22{0})'.format(i)
    ax.text(0.2, 0.4, txt, fontsize=24)

plt.show()
```

세자리수 정수가 아닌 세개의 정수로 지정하는 방법, -> 콤마 단락기법]

```python
fig = plt.figure()

# 서브를롯 작성
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

plt.show()
```

subplot() 함수를 이용하여 서브플롯 배치하기
- 좀더 간결하게 코드 기술 가능

```python
#  피겨 오브젝트 작성과 서브플롯 배치를 동시에 실행
fig, axes = plt.subplots(2, 2)
```
