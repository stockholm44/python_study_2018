# 05-03_색지정과 컬러맵

색과 불투명도 설정하기
- color 채우기색
- edgecolor 테두리색
- alpha 불투명도
- framealpha 범례의 불투명도

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 6, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111)

# 그리기
ax.scatter(x, y, label='sin', color='lightblue', edgecolor='blue')
ax.scatter(x, z, label='cos', color='lightblue', edgecolor='blue', alpha=0.3)
ax.legend(framealpha=1, loc='lower right')

plt.show()
```

색지정하기
1. Matplotlib 컬러 스트링 : 'b','g','r','c','m','y','k','w'
2. RGB 컬러코드 : (270,130,180)
3. RGBa 컬러코드 : (255,215,0,1)
4. 16진수 컬러코드 : "#4682b4, #ffd700"
5. 그레이스케일 : 0.75
6. HTML 컬러이름 : 'red','blue', 'green'
7. X11/CSS4 컬러이름 : 'Red', 'Blue', 'Green'
8. xkcd 컬러 조사이름 : 'xkcd:sky blue'
9. 정규표현 컬러 : 'C0','C1','C2','C3','C4','C5','C6','C7','C8','C9'
10. Tableau 컬러 : 'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'

```python
# 색지정 예
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(111)

x = np.arange(0,3,0.1)
y = np.cos(x)

#여러가지 방법으로 색지정해서 그리기
ax.scatter(x, y, label="color string", color="b")
ax.scatter(x + 0.5, y, label="RGBA tuple", color=(0.1, 0, 0.5, 1))
ax.scatter(x + 1.0, y, label="RGBA 16", color="#330099")
ax.scatter(x + 1.5, y, label="Gray scale", color="0.3")
ax.scatter(x + 2.0, y, label="HTML color", color="DarkOliveGreen")
ax.scatter(x + 2.5, y, label="X11/CSS", color="DarkSeaGreen")
ax.scatter(x + 3.0, y, label="Regular expression", color="C8")
ax.scatter(x + 3.5, y, label="Tableau", color="tab:gray")
ax.legend()
plt.show()
```

컬러맵 이용하기.
- matplotlib.cm 모듈 사용

```python
import matplotlib.cm as cm

fig = plt.figure()
ax = fig.add_subplot(111)

x = np.arange(20)
y = np.arange(20)

ax.scatter(x, y + 5, color=cm.spring(0.5), label='sprint(0.5)')
ax.scatter(x, y, color=cm.get_cmap('winter')(0.5), label='winter(0.5)')

ax.legend()

plt.show()
```

단계적으로 색을 변화시킬 경우

```python
fig = plt.figure()
ax = fig.add_subplot(111)

n = 20
x = np.arange(n)
y = np.ones(n)

ax.scatter(x, y, c=x, cmap=cm.Set1)

plt.show()
```

개별로 색을 지정한 산포도 그래프

```python
fig = plt.figure()
ax = fig.add_subplot(111)

n = 20
x = np.arange(n)
y = np.ones(n)

for i in range(n):
    color = cm.Set1(i / n * 0.5)
    ax.scatter(x[i], y[i], c=color)
```
