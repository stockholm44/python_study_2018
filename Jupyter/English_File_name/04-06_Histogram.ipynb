# 4-6 히스토그램

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# 데이터 세트 작성
mu = 100 # 평균값
sigma = 10 # 표준편차
np.random.seed(0)
x = np.random.normal(mu, sigma, 10000)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(x)

plt.show()
```

막대의 폭과 수를 변경하는 경우
- rwidth : 막대의 폭
- bins : 막대의 갯수

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(x, rwidth = 0.9, bins = 16)

plt.show()
```

히스토그램 활용하기

```python
import os
import pandas as pd

base_url = 'https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/'
anime_master_csv = os.path.join(base_url, 'anime_master.csv')

df = pd.read_csv(anime_master_csv, index_col='anime_id')
df.head()
```

```python
# 평점 시각화
fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(df['rating'], range(0, 10), rwidth=0.9)
ax.set_title('Rating')

plt.show()
```

```python
# 에피소드 수의 시각화
fig = plt.figure()
ax = fig.add_subplot(111)
df_tv = df[df['type']=='TV']

ax.hist(df_tv['episodes'], rwidth=0.9)
ax.set_title('Episodes')

plt.show()
```

```python
# 히스토그램 범위 지정
fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(df_tv['episodes'], rwidth=0.9, range=(0,100))
ax.set_title('Episodes(0-100)')

plt.show()
```

수평 히스토그램
- orientation = horizontal 설정시 가능

```python
np.random.seed(0)
x = np.random.normal(100,10,10000)
fig = plt.figure()
ax = fig.add_subplot(111)

# orientation을 horizontal로 설정
ax.hist(x, rwidth=0.9, bins=16, orientation='horizontal')

plt.show()
```

상대도수 히스토그램
- 데이터수가 다르그룹의 히스토그램을 비교하는 경우에는 상대도수 이용하는게 비교에 용이

- 상대도수 총합이 1이됨.

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# normed=True로 지정
ax.hist(df['rating'], density=True, rwidth=0.9)

plt.show()
```

누적히스토그램(누적 도수 그림)
- cumulative=True 지정

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# cumulative를 True로 지정
ax.hist(df['rating'], cumulative=True, rwidth=0.9, density=True)

plt.show()
```

계급 폭 지정

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(df['rating'], bins = [2, 4, 5.5, 6.5, 7, 7.5, 8.5, 10], rwidth = 0.9)

plt.show()
```

근사 곡선 추가
- 근사곡선은 히스토그램을 그린 후에 꺾은선 그래프로 그린다.
- 여기에서는 정규분포를 이용해서 근사한다.

```python
bins = 50 # 막대수
dfmin = np.min(df['rating']) # 데이터 최솟값
dfmax = np.max(df['rating']) # 데이터 최댓값

# 히스토그램 그리기
fig = plt.figure()
ax = fig.add_subplot(111)

ax.hist(df['rating'], bins = bins, range = (dfmin, dfmax), normed = True, rwidth = 0.9)

# 평균과 표준편차
mu, sigma = df['rating'].mean(), df['rating'].std()

# X값
x = np.linspace(dfmin, dfmax, bins) # 막대 단락 값

# 근사적 확률밀도함수를 이용해 Y값 생성
y = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-(x - mu)**2/ (2 * sigma**2))

# 근사곡선 그리기
ax.plot(x, y)

plt.show()
```

여러 그룹을 겹쳐서 그리기
- 같은 서브플랏에 반복해서 그리면 겹쳐서 그리기 가능!

```python
types = df['type'].unique()
labels = types.tolist()

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

b_num = np.arange(0, 10.5, 0.5)

for t in types:
    ax.hist(df.loc[df['type']==t, 'rating'],
    bins = b_num,
    rwidth = 0.9,
    alpha = 0.5,
    label = t)

ax.legend()
ax.set_xlabel('rating')
ax.set_ylabel('Count(rating)')

plt.show()
```

여러 그룹을 나열하여 그리기

```python
# 데이터세트 작성
dataset = [df.loc[df['type'] == t, 'rating'] for t in types]

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.hist(dataset,
        bins=np.arange(0,10.5,0.5),
        rwidth = 0.9,
        alpha=0.8,
        label=labels)
ax.legend()
ax.set_xlabel('rating')
ax.set_ylabel('Count(rating)')

plt.show()
```

여러그룹을 누적해서 그리기

```python
# datasets, labels는 나열해서 그리기로 작성한 것을 사용
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.hist(dataset,
        bins=np.arange(0,10.5,0.5),
        rwidth=0.9,
        alpha=0.7,
        label=labels,
        stacked=True)

ax.legend()
ax.set_xlabel('rating')
ax.set_ylabel('Count(rating)')

plt.show()
```
