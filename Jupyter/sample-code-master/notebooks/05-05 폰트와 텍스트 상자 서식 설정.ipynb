# 05-05 폰트와 텍스트 상자 서식 설정

폰트 서식
- matplotlib.font_manager.FontProperties
    - family : 폰트 패밀리 종류
    - style : 폰트 스타일('normal', 'italic', 'oblique')
    - weight : 폰트 굵기(0~1000, 'ultralight', 'light', 'normal', 'regular','book',' medium','roman','semibold','demibold','demi','heavy','extra bold','black')
    - size : 폰트사이즈(수치,'xx-small','x-small','small','medium','large','x-large','xx-large','larger','smaller')
    - alpha : 불투명도 (0~1)
    - color : 폰트 색

텍스트 상자 서식
- matplotlib.text.Text
    - bbox : 텍스트 박스 서식 설정
    - verticalalignment : 수직방향 배치('center', 'top','bottom','baseline')
    - horizontalalignment : 수평방향 배치('center','right','left')
    - multialignment : 복수형 배치('center','right','left')
    - linespacing : 행간(폰트사이즈에 대한 비율
    - fontproperties : 폰트 서식
    - rotation : 회전(각도, 'vertical','horizontal'
    - rotation_mode : 회전방법('anchor')
    - backgroundcolor : 배경색

```python
import os
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt, font_manager
from matplotlib import pyplot as plt, rcParams

#여기에서 지정된 설정은 초기 설정 값으로 이용된다
rcParams['font.sans-serif'] = 'Source Han Sans K'

# #폰트 캐쉬를 재구축한다. 
# font_manager._rebuild()

# if os.name == 'nt':
#     font_dir = font_manager.win32FontDirectory()
# else:
#     font_dir = '/Users/pydata/Library/Fonts'

# font_path = os.path.join(font_dir, 'SourceHanSansK-Regular.otf')
# font = font_manager.FontProperties(fname=font_path, size=14)
```

```python
# 텍스트 그리기

fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

# 텍스트 그리기
ax.text(2,4, 'Text', size=40)

ax.set_xticks([i * 2 for i in range(6)])
ax.set_yticks([i * 2 for i in range(6)])

ax.grid(linestyle='-') # 눈금선 그리기

plt.show()
```

폰트 서식 설정하기

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

# 폰트 설정하고 텍스트 그리기
ax.text(0.2, 0.6, 'Text', alpha=0.5, color='Red', size=40)

# 폰트 설정하고 텍스트 그리기(사전 사용)
fps={
    'family': 'fantasy',
    'style': 'italic',
    'weight': 'heavy',
    'size':40,
    'color': 'Green',
    }

ax.text(0.2, 0.2, 'Text', alpha=0.5, color='Red', fontdict=fps)

ticks = [0.2 * i for i in range(6)]
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.grid(linestyle='-')

plt.show()
```

텍스트상자 서식 설정하기
- 인수 bbox에서 설정

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

# 텍스트 상자 서식 사전
boxprops = {
    'facecolor': 'pink',
    'edgecolor': 'red',
    'alpha': 0.5,
    'boxstyle': 'Sawtooth',
    'linewidth': 2
    }

# 텍스트 상자 그리기
ax.text(0.07, 0.5, '텍스트 상자\n서식 사전에 따른 \n설정', bbox=boxprops, size=18)
ax.text(0.07, 0.2, '채우기색 설정', bbox={'alpha':0.2}, backgroundcolor='Blue', size=16)

ticks = [0.2 * i for i in range(6)]
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.grid(linestyle='-')

plt.show()
```

텍스트상자의 수평 방향 배치 설정하기
- horizontalalignment에서 설정
- left, center, right

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

x, y = [0.5, 0.5], [0, 4]
ax.plot(x, y)

# 텍스트 상자 그리기(수평방향 배치)
ax.text(0.5, 3, 'left', horizontalalignment='left', size=20)
ax.text(0.5, 2, 'center', horizontalalignment='center', size=20)
ax.text(0.5, 1, 'right', horizontalalignment='right', size=20)

ax.set_yticks([i for i in range(5)])
ax.set_xticks([0.5 * i for i in range(3)])

plt.show()
```

텍스트 상자의 수직방향 배치 설정하기
- verticalalignment 로 설정
    - bottom : 좌표에 텍스트 박스 왼쪽 하단 모서리를 맞춤.
    - baseline : 좌표에 문자열 왼쪽 하단 모서리를 맞춤.
    - center : 텍스트박스 왼쪽 모서리 - Y 축방향을 맞춤
    - top : 텍스트 박스 왼쪽 상단 모서리를 맞춘다.

```python
fig = plt.figure(figsize=(8,3))
ax = fig.add_subplot(111)

# 베이스 라인 그리기
x, y = [0, 9], [0.5, 0.5]
ax.plot(x,y)

# 텍스트 박스 그리기 (수직 방향 배치)
ax.text(0, 0.5, 'bottom', verticalalignment='bottom', size=20)
ax.text(2.5, 0.5, 'baseline', verticalalignment='baseline', size=20)
ax.text(5.5, 0.5, 'center', verticalalignment='center', size=20)
ax.text(8, 0.5, 'top', verticalalignment='top', size=20)

ax.set_xticks([i * 3 for i in range(4)])
ax.set_yticks([0.5 * i for i in range(3)])

plt.show()
```

텍스트상자 안 문자열의 수평방향 배치 설정하기
- multialignment로 설정

```python
fig = plt.figure(figsize=(4, 7))
ax = fig.add_subplot(111, xlim=[0,5], ylim=[0,14])

# 텍스트 상자 안 문자열의 수평방향 배치를 설정하고 그리기
ax.text(1,10, '왼쪽 정렬\nLeft\nalignment', multialignment='left', size=20)
ax.text(1,6, '가운데 정렬\nCenter\nalignment', multialignment='center', size=20)
ax.text(1,2, '오른쪽 정렬\nRight\nalignment', multialignment='right', size=20)

ax.grid(linestyle='-')

plt.show()
```

텍스트상자 안 문자열의 줄 간격 설정하기
- linespacing으로 설정

```python
fig = plt.figure(figsize=(3.5,5))
ax = fig.add_subplot(111, xlim=[0,4], ylim=[0,7])

# 줄 간격 설정
ax.text(1,5, '줄간격 1\nspace 1', linespacing=1, size=20)
ax.text(1,3, '줄간격 1.5\nspace 1.5', linespacing=1.5, size=20)
ax.text(1,1, '줄간격 2\nspace 2', linespacing=2, size=20)

ax.grid(linestyle='-')

plt.show()
```

텍스트 상자 회전하기
- rotation과 rotation_mode 로 설정

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

# 베이스 라인 그리기
axes = [ax1, ax2, ax3, ax4]
x, y = [0.5, 0.5], [0, 1]

for ax in axes:
    ax.plot(x, y, color='SteelBlue')
    ax.plot(y, x, color='SteelBlue')
    
# 서브플롯 1~3 그리기
ax1.text(0.5, 0.5, "horizontal", rotation="horizontal", size=12)
ax2.text(0.5, 0.5, "vertical", rotation="vertical", size=12)
ax3.text(
    0.5,
    0.5,
    "rotation=45\nwith anchor\nsetting",
    rotation=45,
    rotation_mode="anchor",
    size=12,
)
# 서브플롯 4 그리기
# 배치비교용 텍스트 상자 그리기
boxprops = {"facecolor": "Pink", "edgecolor": "Pink", "alpha": 0.5}
ax4.text(
    0.5,
    0.5,
    "rotation=45\nwith anchor\nsetting",
    rotation=0,
    bbox=boxprops,
    color="white",
    size=12,
)
# 택스트 그리기
ax1.text(0, 0.9, "1)", size=16)
ax2.text(0, 0.9, "2)", size=16)
ax3.text(0, 0.9, "3)", size=16)
ax4.text(0, 0.9, "4)", size=16)
# 회전한 텍스트 상자 그리기
boxprops = {"facecolor": "None", "edgecolor": "Pink"}
ax4.text(0.5, 0.5, "rotation=45\nwithoutanchor\nsetting", bbox=boxprops, rotation=45, size=12)
for ax in axes:
    ax.set_xticks([0.5 * i for i in range(3)])
    ax.set_yticks([0.5 * i for i in range(3)])
plt.show()
```
