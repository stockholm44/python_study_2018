제너레이터와 yield 키워드를 사용하여 적은 코드로 이터레이터를 더 빨리 작성해보자.

무한 제너레이터

```python
# 기존 이터레이터 (7줄이나 된다.)
class Repeater:
    def __init__(self, value):
        self.value = value
    
    def __iter__(self):
        return self
    
    def __next__(self):
        return self.value
```

```python
# 제너레이터로 작성시 
def repeater(value):
    while True:
        yield value
```

```python
for x in repeater('HI'):
    print(x)
```

```python
repeater('Hey')
```

```python
# 제너레이터 함수의 코드는 제너레이터 객체에서 next()가 호출될 때만 실행된다.
generator_obj = repeater('Hey')
next(generator_obj)
```

함수안에서 return 문이 실행되면 함수의 호출자에게 제어권을 영원히 넘겨버린다.  
반면 yield가 호출되면 함수의 호출자에게 제어권을 다시 전달하지만 '일시적으로만' 한다.  


return 문이 함수의 로컬 상택를 삭제하는 반면 yield 문은 함수를 일시 중단하고 로컬상태를 유지.  
실질적으로 이것은 지역변수와 제너레이터 함수 실행 상태가 잠시 숨겨질 뿐 완전히 사라지지 않는다는 것을 의미함.

```python
iterator = repeater('Hi')
next(iterator)
```

```python
next(iterator)
```

```python
next(iterator)
```

생성을 멈추는 제너레이터

```python
def repeat_three_times(value):
    yield value
    yield value
    yield value
```

```python
for x in repeat_three_times('Hey there'):
    print(x) # 마지막에는 StopIteration 예외를 발생시켰을것으로 가정하지만 표현이 안되있다. 
```

```python
# next문으로 StopIteration 확인
iterator = repeat_three_times('Hey there')
next(iterator)
```

```python
# 횟수지정 BoundedRepeater 클래스
class BoundedRepeater:
    def __init__(self, value, max_repeats):
        self.value = value
        self.max_repeats = max_repeats
        self.count = 0
        
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.count >= self.max_repeats:
            raise StopIteration
        self.count += 1
        return self.value
```

```python
# 횟수지정 bounded_repeater 제너레이터
def bounded_repeater(value, max_repeats):
    count = 0
    while True:
        if count >= max_repeats:
            return
        count += 1
        yield value
```

```python
for x in bounded_repeater('Hi',5):
    print(x)
```

```python
def bounded_repeater(value, max_repeats):
    for i in range(max_repeats):
        yield value
```

```python
for x in bounded_repeater('Hi',5):
    print(x)
```
